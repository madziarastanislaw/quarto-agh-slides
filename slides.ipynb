## Slide with an equation

Any piecewise-continuous $2\pi$-periodic function can be approximated by a Fourier series:

$$
f(x)=\sum\limits_{n=-\infty}^{\infty} c_n e^{inx}
$$

where

$$
c_n=\frac{1}{2\pi}\int\limits_{-\pi}^{\pi} f(x)e^{-inx}\,dx
$$
